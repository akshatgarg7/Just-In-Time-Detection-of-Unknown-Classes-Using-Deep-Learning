```python
import pickle

# open the file in binary read mode
with open('lists.pkl', 'rb') as f:
    # unpickle the two lists
    same_disssimilarity = pickle.load(f)
    different_disssimilarity = pickle.load(f)

# print the unpickled lists
print(same_disssimilarity)
print(different_disssimilarity)
```

```python
from utils.other_utils import threshold
```

```python
my_list = [round(x * 0.01, 2) for x in range(-300, 301)]
# my_list
```

```python
final = []
same = []
different = []
for i in my_list:
    a,b,c = threshold(same_disssimilarity,different_disssimilarity,i)
    final.append(len(a))
    same.append(len(b))
    different.append(len(c))
```

```python
import plotly.express as px

# create the line plot
fig = px.line(x=my_list, y=final)

# set the x-axis range to -1 to 1 and y-axis range to -100 to 2500
fig.update_xaxes(range=[-1, 2])
fig.update_yaxes(range=[-100, 2500])

# add a vertical line at x=0.2
fig.add_shape(type='line', x0=0.2, x1=0.2,
              y0=0, y1=max(final),
              line=dict(color='red', width=1, dash='dash'),
              name='Minimum Number of Misclassification')

fig.update_layout(
    title={
        'text': 'Total missclassification vs threshold',
        'x': 0.5,
        'xanchor': 'center'
    },
    xaxis_title='Threshold',
    yaxis_title='Missclassified images',
    xaxis=dict(showgrid=True, gridwidth=1, gridcolor='DarkGray', linecolor='DarkGray', linewidth=2),
    yaxis=dict(showgrid=True, gridwidth=1, gridcolor='DarkGray', linecolor='DarkGray', linewidth=2),
    showlegend=True
)
# display the plot
fig.show()
```

```python
import plotly.graph_objects as go

# create trace for the first line
trace1 = go.Scatter(x=my_list, y=same, mode='lines', name='Same class Images')

# create trace for the second line
trace2 = go.Scatter(x=my_list, y=different, mode='lines', name='Different class Images')

# create the figure object and add both traces
fig = go.Figure()
fig.add_trace(trace1)
fig.add_trace(trace2)

# set the x-axis range to -1 to 1 and y-axis range to -100 to 2500
fig.update_xaxes(range=[-0.5, 1.5])
fig.update_yaxes(range=[-100, 1500])

# find the intersection point of the two lines
intersection_point = my_list[same.index(max(same))]

# add a vertical line at x=0.2
fig.add_shape(type='line', x0=0.19, x1=0.19,
              y0=0, y1=max(final),
              line=dict(color='red', width=1, dash='dash'),
              name='Intersection of both')


fig.update_layout(
    title={
        'text': 'Optimal Threshold Value for Minimum Misclassification',
        'x': 0.5,
        'xanchor': 'center'
    },
    xaxis_title='Threshold',
    yaxis_title='Missclassified images',
    xaxis=dict(showgrid=True, gridwidth=1, gridcolor='DarkGray', linecolor='DarkGray', linewidth=2),
    yaxis=dict(showgrid=True, gridwidth=1, gridcolor='DarkGray', linecolor='DarkGray', linewidth=2),
    showlegend=True
)

# display the plot
fig.show()
```

```python
import pandas as pd
import plotly.graph_objs as go


df = pd.DataFrame(
    dict(
        year=['Experiment 1', 'Experiment 2', 'Experiment 3'],
        var1=[260/22, 7613/22, 2502/22],
        var3=[0,94,142],
        var4=[25,32,59],
    )
)
        
fig = go.Figure()

fig.update_layout(
    title={
        'text': 'Results on Three Different Experiments of Rotation',
        'x': 0.5,
        'xanchor': 'center'
    },
    template="simple_white",
    xaxis=dict(title_text="Experiment"),
    yaxis=dict(title_text="Value"),
    barmode="stack",
)

groups = ['var1','var3','var4']
colors = ["blue","green","red"]
names = ['Number of Intersection per 1K Image Pairs','Same Images Misclassified as Different Images','Different Images Misclassified as Same Images']

i = 0
for r, n, c in zip(groups, names, colors):
    ## put var1 and var2 together on the first subgrouped bar
    if i == 0:
        fig.add_trace(
            go.Bar(x=[df.year, ['Number of Intersection']*len(df.year)], y=df[r], name=n, marker_color=c),
        )
    ## put var3 and var4 together on the first subgrouped bar
    else:
        fig.add_trace(
            go.Bar(x=[df.year, ['Number of Misclassification']*len(df.year)], y=df[r], name=n, marker_color=c),
        )
    i+=1

fig.show()   
```
